```python
x=np.linspace(0,5,11)
x
```

```python
plt.plot(x,x)
plt.show()
```

```python
plt.plot(x,x**2)
plt.xlabel("x")
plt.ylabel("x**2")
plt.title("test",color="red")
```

```python
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")
plt.legend()
plt.show()
```

```python
         #SUBPLOT
```

```python
plt.subplot(1,3,1) #(no.of row,no.of column,plot you want to work on)
plt.plot(x,x)
plt.subplot(1,3,2)
plt.plot(x,x**2)
plt.subplot(1,3,3)
plt.plot(x,x**3)
plt.show()
```

```python
plt.subplot(1,3,1) #(no.of row,no.of column,plot you want to work on)
plt.plot(x,x,"g^",) #g^ uppertriangle with green colour
plt.xlabel("x1")
plt.ylabel("y1")
plt.subplot(1,3,2)
plt.plot(x,x**2,"rs--")
plt.xlabel("x2")
plt.ylabel("y2")
plt.subplot(1,3,3)
plt.plot(x,x**3,"c:")
plt.xlabel("x3")
plt.ylabel("y3")
plt.tight_layout() #to avoid graphs overlapping each other
plt.show()
```

```python
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic",color="red",marker="s",mec="green",mfc="yellow",mew=3,lw=5)
plt.legend()
plt.show()
```

```python
fig=plt.figure()
axs=fig.add_axes([.1,.1,.8,.8])
axs.set_xlabel("X LABEL")
axs.set_ylabel("Y LABEL")
axs.set_title("SAMPLE PLOT")
axs1=fig.add_axes([.2,.5,.4,.3])
axs1.set_xlabel("X LABEL")
axs1.set_ylabel("Y LABEL")
axs1.set_title("inset plot")
axs.plot(x,x**2,color="red")
axs1.plot(x**3,x,"gs")
```

```python
fig,ax=plt.subplots(figsize=(8,3),dpi=300,facecolor="yellow",edgecolor="blue") #in inches
ax.plot(x,x**3,label="x^3")
ax.plot(x,x**2,label="x^2")
ax.legend(loc=0) #loc=0 choose best position to place legend
ax.set_title("SAMPLE PLOT",color="green",fontsize=20)
```

```python
fig,ax=plt.subplots(figsize=(8,3),dpi=300,facecolor="yellow",edgecolor="blue") #in inches
ax.plot(x,x**3,label="x^3")
ax.plot(x,x**2,label="x^2")
ax.legend(loc=(.2,.2))
ax.set_title("SAMPLE PLOT",color="green",fontsize=20)
```

```python
fig,ax=plt.subplots(figsize=(8,3),dpi=300,facecolor="yellow",edgecolor="blue") #in inches
ax.plot(x,x**3,label="x^3")
ax.plot(x,x**2,label="x^2")
ax.legend(loc=1) #rightmost side
ax.axis("tight")
ax.set_title("SAMPLE PLOT",color="green",fontsize=20)
```

```python
fig1,axes=plt.subplots(1,3,figsize=(6,4))
i=1
for ax in axes:
    ax.plot(x,x**i)
    i+=1
axes[2].set_xlim(0,2) #x axis from 0 to 2 in fig3
axes[2].set_ylim(0,10) # y axis from 0 to 10 in fig3
fig.tight_layout()  #to remove overlapping  
plt.show()
```

```python
x1=np.random.normal(size=100)
y1=np.random.normal(size=len(x1))
s1=np.random.standard_normal(100) #size array
c1=np.random.standard_normal(100) # for colour
```

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(x1,y1)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(x1,y1,s=s1*50)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(x1,y1,s=s1*50,c=c1)
plt.show()
```

```python
xx=np.linspace(0,8,16)
yy=np.random.uniform(0,5,len(xx))
yy1=1.5+yy
fig,ax=plt.subplots()
ax.fill_between(xx,yy,yy1,alpha=0.4)#fill inbetween the two graphs #alpha diminishes opacity
ax.plot(xx,(yy+yy1)/2,"r:") #line
```
