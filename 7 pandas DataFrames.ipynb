```python
#Merging dataframes #JOINS
```

```python
#INNERJOIN   #OUTERJOIN #LEFTINNERJOIN  #RIGHTINNERJOIN
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
right
```

```python
pd.merge(left,right,how="inner",on=["key1","key2"]) #innerjoin
```

```python
pd.merge(left,right,how="outer",on=["key1","key2"]) #outerjoin
```

```python
pd.merge(left,right,how="left",on=["key1","key2"]) #leftouterjoin
```

```python
pd.merge(left,right,how="right",on=["key1","key2"]) #rightouterjoin
```

```python
#INNERJOIN   #OUTERJOIN #LEFTINNERJOIN  #RIGHTINNERJOIN
```

```python
left1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],'B': ['B0', 'B1', 'B2'],index=['K0','K1','K2']})
right1= pd.DataFrame({'C': ['C0', 'C2', 'C3'],'D': ['D0', 'D2', 'D3'],index=['K0','K2','K3']})
```

```python
left1
```

```python
right1
```

```python
left1.join(right1,how="outer")
```

```python
left1.join(right1,how="left")
```

```python
left1.join(right1,how="right")
```
