```python
iris=sns.load_dataset("iris")
```

```python
p=sns.PairGrid(iris)
p.map(sns.scatterplot)
```

```python
p=sns.PairGrid(iris)
p.map_diag(sns.histplot)  #diagonal histogram,resst scatter
p.map_offdiag(sns.scatterplot)
```

```python
p=sns.PairGrid(iris)
p.map_diag(sns.histplot)    #diaginal
p.map_lower(sns.scatterplot)  #upperdiagonal
p.map_upper(sns.kdeplot)     #lower diagonal
```

```python
p=sns.PairGrid(iris,hue="species")
p.map_diag(sns.histplot)    #diaginal
p.map_lower(sns.scatterplot)  #upperdiagonal
p.map_upper(sns.kdeplot)#lower diagonal
p.add_legend()
```
